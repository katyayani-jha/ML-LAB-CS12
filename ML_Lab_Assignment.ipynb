```python
import numpy as np

arr = np.array([1, 2, 3, 6, 4, 5])
reversed_arr = arr[::-1]
print("Reversed array:", reversed_arr)

array1 = np.array([[1, 2, 3], [2, 4, 5], [1, 2, 3]])
flattened_array_method1 = array1.flatten()
flattened_array_method2 = array1.ravel()
print("Flattened array (method 1):", flattened_array_method1)
print("Flattened array (method 2):", flattened_array_method2)

arr1 = np.array([[1, 2], [3, 4]])
arr2 = np.array([[1, 2], [3, 4]])
arrays_equal = np.array_equal(arr1, arr2)
print("Are arr1 and arr2 equal?", arrays_equal)
```

```python
x = np.array([1, 2, 3, 4, 5, 1, 2, 1, 1, 1])
x_unique, x_counts = np.unique(x, return_counts=True)
x_most_frequent_value = x_unique[np.argmax(x_counts)]
x_indices = np.where(x == x_most_frequent_value)[0]

print("Array x:")
print("Most frequent value:", x_most_frequent_value)
print("Indices:", x_indices)

y = np.array([1, 1, 1, 2, 3, 4, 2, 4, 3, 3])
y_unique, y_counts = np.unique(y, return_counts=True)
y_most_frequent_value = y_unique[np.argmax(y_counts)]
y_indices = np.where(y == y_most_frequent_value)[0]

print("\nArray y:")
print("Most frequent value:", y_most_frequent_value)
print("Indices:", y_indices)
```

```python
import numpy as np

gfg = np.matrix('[4, 1, 9; 12, 3, 1; 4, 5, 6]')

sum_all_elements = gfg.sum()

# Row-wise
sum_row_wise = gfg.sum(axis=1)

# Column-wise
sum_column_wise = gfg.sum(axis=0)

print("Sum of all elements:", sum_all_elements)
print("Sum of all elements row-wise:", sum_row_wise)
print("Sum of all elements column-wise:", sum_column_wise)
```

```python
import numpy as np

n_array = np.array([[55, 25, 15], [30, 44, 2], [11, 45, 77]])

# Sum of diagonal elements
sum_diagonal = np.trace(n_array)

# Eigenvalues and eigenvectors of matrix
eigenvalues, eigenvectors = np.linalg.eig(n_array)

# Inverse of matrix
inverse_matrix = np.linalg.inv(n_array)

# Determinant of matrix
determinant = np.linalg.det(n_array)

print("Sum of diagonal elements:", sum_diagonal)
print("Eigenvalues of the matrix:", eigenvalues)
print("Eigenvectors of the matrix:\n", eigenvectors)
print("Inverse of the matrix:\n", inverse_matrix)
print("Determinant of the matrix:", determinant)
```

```python
import numpy as np

p1 = np.array([[1, 2], [2, 3]])
q1 = np.array([[4, 5], [6, 7]])

product1 = np.matmul(p1, q1)
covariance1 = np.cov(p1.flatten(), q1.flatten())

print("Case i:")
print("Product of p1 and q1:")
print(product1)
print("Covariance matrix between p1 and q1:")
print(covariance1)


p2 = np.array([[1, 2], [2, 3], [4, 5]])
q2 = np.array([[4, 5, 1], [6, 7, 2]])

product2 = np.matmul(p2, q2)
covariance2 = np.cov(p2.flatten(), q2.flatten())

print("\nCase ii:")
print("Product of p2 and q2:")
print(product2)
print("Covariance matrix between p2 and q2:")
print(covariance2)
```

```python
import numpy as np
from scipy.spatial import distance

x = np.array([[2, 3, 4], [3, 2, 9]])
y = np.array([[1, 5, 0], [5, 10, 3]])

inner_product = np.dot(x, y.T)

outer_products = []
for row_x in x:
    for row_y in y:
        outer_product = np.outer(row_x, row_y)
        outer_products.append(outer_product)

# Using broadcasting to compute the Cartesian product of two matrices
cartesian_product = np.array(np.meshgrid(x.flatten(), y.flatten())).T.reshape(-1, 2)

print("Inner Product:")
print(inner_product)

print("\nOuter Products:")
for i, outer_product in enumerate(outer_products):
    print(f"Outer Product {i+1}:")
    print(outer_product)

print("\nCartesian Product:")
print(cartesian_product)
```

2 Questions based on NumPy Mathematics & Statistics

```python
import numpy as np

array = np.array([[1, -2, 3], [-4, 5, -6]])

# Finding element-wise absolute value
absolute_values = np.abs(array)

# Finding the 25th, 50th, and 75th percentile of the flattened array, for each column, and for each row

flattened_array = array.flatten()

percentiles_flattened = {
    '25th': np.percentile(flattened_array, 25),
    '50th': np.percentile(flattened_array, 50),
    '75th': np.percentile(flattened_array, 75)
}

percentiles_columns = {
    '25th': np.percentile(array, 25, axis=0),
    '50th': np.percentile(array, 50, axis=0),
    '75th': np.percentile(array, 75, axis=0)
}

percentiles_rows = {
    '25th': np.percentile(array, 25, axis=1),
    '50th': np.percentile(array, 50, axis=1),
    '75th': np.percentile(array, 75, axis=1)
}

# Mean, Median, and Standard Deviation of the flattened array

mean_flattened = np.mean(flattened_array)
median_flattened = np.median(flattened_array)
std_dev_flattened = np.std(flattened_array)

# Mean, Median, and Standard Deviation for each column

mean_columns = np.mean(array, axis=0)
median_columns = np.median(array, axis=0)
std_dev_columns = np.std(array, axis=0)

# Mean, Median, and Standard Deviation for each row

mean_rows = np.mean(array, axis=1)
median_rows = np.median(array, axis=1)
std_dev_rows = np.std(array, axis=1)

print("Element-wise Absolute Values:")
print(absolute_values)

print("\nPercentiles of Flattened Array:")
print(percentiles_flattened)

print("\nPercentiles for Each Column:")
print(percentiles_columns)

print("\nPercentiles for Each Row:")
print(percentiles_rows)

print("\nMean, Median, and Standard Deviation of Flattened Array:")
print(f"Mean: {mean_flattened}")
print(f"Median: {median_flattened}")
print(f"Standard Deviation: {std_dev_flattened}")

print("\nMean, Median, and Standard Deviation for Each Column:")
print(f"Mean: {mean_columns}")
print(f"Median: {median_columns}")
print(f"Standard Deviation: {std_dev_columns}")

print("\nMean, Median, and Standard Deviation for Each Row:")
print(f"Mean: {mean_rows}")
print(f"Median: {median_rows}")
print(f"Standard Deviation: {std_dev_rows}")
```

```python
import numpy as np

a = np.array([-1.8, -1.6, -0.5, 0.5, 1.6, 1.8, 3.0])

floor_values = np.floor(a)

ceiling_values = np.ceil(a)

truncated_values = np.trunc(a)

rounded_values = np.round(a)

print("Original Array:")
print(a)

print("\nFloor Values:")
print(floor_values)

print("\nCeiling Values:")
print(ceiling_values)

print("\nTruncated Values:")
print(truncated_values)

print("\nRounded Values:")
print(rounded_values)
```

3 Questions based on Searching & Sorting

```python
import numpy as np

array = np.array([10, 52, 62, 16, 16, 54, 453])

sorted_array = np.sort(array)

sorted_indices = np.argsort(array)

four_smallest = np.partition(array, 4)[:4]

five_largest = np.partition(array, -5)[-5:]

print("Original Array:")
print(array)

print("\nSorted Array:")
print(sorted_array)

print("\nIndices of Sorted Array:")
print(sorted_indices)

print("\n4 Smallest Elements:")
print(four_smallest)

print("\n5 Largest Elements:")
print(five_largest)
```

```python
import numpy as np

array = np.array([1.0, 1.2, 2.2, 2.0, 3.0, 2.0])

integer_elements = array[np.isclose(array, np.round(array))]

float_elements = array[~np.isclose(array, np.round(array))]

print("Original Array:")
print(array)

print("\nInteger Elements Only:")
print(integer_elements)

print("\nFloat Elements Only:")
print(float_elements)
```
